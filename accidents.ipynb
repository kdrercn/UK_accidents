### Verileri birleştir

```python
import pandas as pd
```

```python
merged_data = pd.concat([ pd.read_csv("data/accidents_2005_to_2007.csv") ,
                         pd.read_csv("data/accidents_2009_to_2011.csv"), 
                         pd.read_csv("data/accidents_2012_to_2014.csv")], axis=0, ignore_index=True)


merged_data=merged_data.reset_index(drop=True)
merged_data
```

#### Null değerler

```python
merged_data.isnull().sum()
```

#### Gereksiz sütunları siliyoruz. 
##### Junction Detail Sütununda hiçbir veri olmadığı için siliyoruz.
##### LSOA_of_Accident_Location ve Did_Police_Officer_Attend_Scene_of_Accident sütunlarını gereksiz olduğu için siliyoruz.

```python
merged_data = merged_data.drop(["Junction_Detail", "Did_Police_Officer_Attend_Scene_of_Accident", "LSOA_of_Accident_Location"], axis=1)
```

#### Drop Duplicates ile birbirinin aynısı verileri silerek veriyi unique hale getirdik.

```python
unique_data = merged_data.drop_duplicates()
```

```python
len(unique_data)
```

```python
unique_data.isnull().sum()
```

#### Verimizde hala çok sayıda null veri taşıyan sütunlar var.

```python
unique_data["Junction_Control"].fillna("Not Specified", inplace = True)
unique_data["Special_Conditions_at_Site"].fillna("No Special Conditions or Not Specified", inplace = True)
unique_data["Carriageway_Hazards"].fillna("No Carriageway Hazards or Not Specified", inplace = True)
unique_data["Road_Surface_Conditions"].fillna("Not Specified", inplace = True)   
```

```python
unique_data.isnull().sum()
```

### Boş satırları silme

#### Kullanmayacağımız sütünleri sildik. Çok fazla Null içeren değerlere de gerekli değerleri atadıktan sonra az sayıda kalan Null değerli sütunları siliyoruz. 

```python
dropped = unique_data.dropna()
```

#### Null veri çok fazla olan sütunları kaldırıyoruz.

```python
dropped.isnull().sum()
```

#### Son verimizde 1469609 veri kaldı.

```python
len(dropped)
```

```python
clean_data = dropped
```
