```python
merged_data = pd.concat([ pd.read_csv("data/accidents_2005_to_2007.csv") ,
                         pd.read_csv("data/accidents_2009_to_2011.csv"), 
                         pd.read_csv("data/accidents_2012_to_2014.csv")], axis=0, ignore_index=True)


merged_data=merged_data.reset_index(drop=True)
merged_data
```

#### Gereksiz s√ºtunlarƒ± siliyoruz. 
##### Junction Detail S√ºtununda hi√ßbir veri olmadƒ±ƒüƒ± i√ßin siliyoruz.
##### LSOA_of_Accident_Location ve Did_Police_Officer_Attend_Scene_of_Accident s√ºtunlarƒ±nƒ± gereksiz olduƒüu i√ßin siliyoruz.

```python
merged_data = merged_data.drop(["Junction_Detail", "Did_Police_Officer_Attend_Scene_of_Accident", "LSOA_of_Accident_Location"], axis=1)
```

#### Drop Duplicates ile birbirinin aynƒ±sƒ± verileri silerek veriyi unique hale getirdik.

```python
unique_data = merged_data.drop_duplicates()
```

```python
len(unique_data)
```

```python
unique_data.isnull().sum()
```

#### Verimizde hala √ßok sayƒ±da null veri ta≈üƒ±yan s√ºtunlar var.

```python
unique_data["Junction_Control"].fillna("Not Specified", inplace = True)
unique_data["Special_Conditions_at_Site"].fillna("No Special Conditions or Not Specified", inplace = True)
unique_data["Carriageway_Hazards"].fillna("No Carriageway Hazards or Not Specified", inplace = True)
unique_data["Road_Surface_Conditions"].fillna("Not Specified", inplace = True)   
```

```python
unique_data.isnull().sum()
```

### Bo≈ü satƒ±rlarƒ± silme

#### Kullanmayacaƒüƒ±mƒ±z s√ºt√ºnleri sildik. √áok fazla Null i√ßeren deƒüerlere de gerekli deƒüerleri atadƒ±ktan sonra az sayƒ±da kalan Null deƒüerli s√ºtunlarƒ± siliyoruz. 

```python
dropped = unique_data.dropna()
```
